```python
#Find the smaller number in a list of three numbers 

a = [ 2, 3, 4]

if (a[0] < a[1]) & (a[0] < a[2]): #Composed statement. & means 'and'
    print('The lowest value is:' , a[0])
else:
    if a[1] < a[2]: #Nested statement
        print('The lowest value is:' , a[1])
    else:
        print('The lowest value is:' , a[2])         
```

```python
#Assigning a letter to a class score

Score = 90

if Score >= 90:
    Letter = 'A'
else:
    if Score >= 80:
        Letter = 'B'
    else: 
        if Score >= 70:
            Letter = 'C'
        else: 
            if Score >= 60:
                Letter = 'D'
            else:
                Letter = 'F'

print('Score of', Score, 'is a:', Letter)           
```

```python
#Assigning a letter to a class score. More elegant

Score = 55

if Score >= 90:
    Letter = 'A'
elif Score >= 80: #Elif is equivalent to else and if together
    Letter = 'B'
elif Score >= 70:
    Letter = 'C'
elif Score >= 60:
    Letter = 'D'
else:
    Letter = 'F'

print('Score of', Score, 'is a:', Letter)
```

## Conditionals: While

```python
i = 1

while i < 6:
    print(i)
    i += 1 #Adding 1 to i
```

```python
i = 0
while i < 10:
    i += 1 #Adding 1 to i
    if i == 7:
        break #Terminate the cycle and move down to the next blok
    print(i)
    
print('While statement was completed')
```

```python
i = 0
while i < 10:
    i += 1 #Adding 1 to i
    if i == 7:
        continue #Stop the cycle and go back to the while 
    print(i)
else: #Not needed  
    print('While statement was completed')
```

## Loops: For

```python
for i in range(6): # i will take values from 0 to 6-1
    print(i)    
```

```python
for i in range(10):
    if i == 7:
        break
    print(i) 
```

```python
for i in range(10):
    if i == 7:
        continue
    print(i) 
```

```python
a = [1, 3, 6, 9]

for i in a: #i will become each object in the list
    print(i)
    
```
